```python
# Graficando sólo parte de los datos, un filtro

lim = 4 # límite para el filtro de los datos
fs = 20 # tamaño de las fuentes

r2d=sqrt(x**2+y**2)
filtro=where( r2d >= lim )

# tamaño del lienzo
figure(figsize=(5,5))

# límites de los ejes
xlim((-15,15))
ylim((-15,15))

# tamaño de las fuentes en los ejes
xticks(fontsize=16)
yticks(fontsize=16)

# nombres de los ejes
ylabel("Y [kpc]", fontsize=fs)
xlabel("X [kpc]", fontsize=fs)

# título de la gráfica
title("Grafica galaxia", fontsize=fs)

# genera gráfica
plot(x[filtro],y[filtro],".",c="red",label="galaxia")
#plot(x,y,".",c="red",label="galaxia")

# activa la layenda
legend(fontsize=fs)

# exporta el gráfico a un archivo
#show()
savefig("grafica.png")
```

# Varios gráficos en un mismo lienzo

```python

angulos = arange(0.0,2*pi+2*pi/40,2*pi/40) # genera arreglo de numpy
fangulos1 = sin(angulos) # arreglo 
fangulos2 = cos(angulos) # arreglo

figure(figsize=(5,5))
fs = 12
title("grafico con lineas",fontsize=fs)
plot(angulos,fangulos1,'-',label="sin(x)",c="green")
plot(angulos,fangulos2,'-o',label="cos(x)",c="red")
grid()
xlabel("X [rad]",fontsize=fs)
ylabel("Funciones",fontsize=fs)
legend(fontsize=fs)
text(3.0,0.75, 'texto solo', style='italic',fontsize=fs)
text(0.0,0.0, 'texto en una caja', style='italic',
     bbox={'facecolor': 'blue', 'alpha': 0.3, 'pad': 5})


show()
```

# Gráficas en ciclos y campos vectoriales 

```python
from math import sqrt

x = []
y = []
vx = []
vy = []
t = []
xo = 0.0
yo = 10.0
vx_ini = 1.0
vy_ini = 0.0
g= 9.8
tTotal = sqrt(2.0*yo/g)

fs = 14

for dt in arange(0,tTotal,0.1):

  t.append(dt)
  vx.append(vx_ini)
  vyaux = vy_ini - g*dt
  vy.append(vyaux)
  x.append(xo + vx_ini*dt) 
  y.append(yo + vyaux*dt)

  figure(figsize=(4,4))
  xlim((-1,2))
  ylim((-1,11))
  xticks(fontsize=16)
  yticks(fontsize=16)
  ylabel("Y [km]", fontsize=fs)
  xlabel("X [km]", fontsize=fs)
  title("Grafica parabólico", fontsize=fs)
  plot(x,y,".",c="red",label="trayectoria")
  quiver(x, y, vx, vy,scale=50)
  legend(fontsize=fs)
  show()
  

```
