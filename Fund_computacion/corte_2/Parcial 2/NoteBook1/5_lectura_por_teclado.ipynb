# Lectura de valores por teclado
Se hace usando la función input() que lee y devuelve la cadena de caracteres que digita en el teclado:

```python
valor_entrada = input()
```

```python
valor_entrada
```

```python
valor_entrada = input()
```

```python
# Aún el usuario ingrese un valor numérico este entra como una cadena de caracteres y no se puede realizar matemáticas con el
valor_entrada 
```

```python
# Para mejorar la experiencia de ususario se debe mostrar un mensaje antes de leer el valor
valor_entrada = input("Ingresa un número: ")
```

```python
# Una cadena de caracteres no se puede operar con números
valor_entrada + 100  
```

```python
valor_entrada = input("Ingresa un número entero: ")
```

## La función int() 

Convierte una cadena de caracteres a un número entero (*casting* a un entero).

```python
# convierte el valor ingresado por el usuario a un entero
valor_entrada = int(valor_entrada)  
```

```python
valor_entrada
```

```python
# Ahora el némero ya se puede operar
valor_entrada + 10  
```

```python
valor_entrada = input("Introduce un número real: ")
```

## Ahora con la función float(), de cadena a flotante.

```python
# convierte el valor ingresado por el ususario a un flotante
valor_entrada = float(valor_entrada) 
```

```python
10 + valor_entrada
```

```python
valor_entrada = float( input("Ingresa un número decimal o entero: ") )
```
