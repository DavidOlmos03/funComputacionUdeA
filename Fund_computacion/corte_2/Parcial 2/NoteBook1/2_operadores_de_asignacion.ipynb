# **Operadores de asignación**
Actúan directamente sobre la variable modificando su valor actual.

```python
n = 0
```

```python
# Suma en asignación
n += 3 
n
```

```python
# Resta en asgnación
n -= 5
n
```

```python
n = 6
# Producto en asignación
n *= 3
n
```

```python
# División en asignación
n /= 2 
n
```

```python
# Módulo en asignación
n %= 3
n
```

```python
# Potencia en asignación
n = 2
n **= 5 
n
```
