# **Los operadores relacionales**
Sirven para comparar dos valores, dependiendo del resultado de la comparación pueden devolver:
- Verdadero (True), si es cierta
- Falso (False), si no es cierta

```python
# Igual que
5 == 3
```

```python
# Diferente a
5 != 3
```

```python
# Mayor que
5 > 3
```

```python
# Menor que
5 < 3
```

```python
# Mayor o igual que
5 >= 3
```

```python
# Menor o igual que
5 <= 3
```
