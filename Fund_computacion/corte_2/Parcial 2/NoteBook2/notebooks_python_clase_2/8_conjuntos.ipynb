### Pertenencia a conjuntos con *in*

```python
grupo = {'Luis','Fernando','Maria'}
grupo
```

```python
'Luis' in grupo
```

```python
'Mario' in grupo
```

```python
'Fernando' not in grupo
```

```python
grupo2
```

### Cast de lista a conjunto y viceversa
Es muy útil transformar listas a conjuntos para borrar los elementos duplicados automáticamente.

```python
l1 = [1,2,3,3,2,1]
l1
```

```python
c1 = set(l1)
c1
```

```python
l1 = list(c1)
l1
```

```python
l2 = [1,2,3,3,2,1]
```

```python
# En una línea
l2 = list( set( l2 ) )
l2
```

### Cast de cadena a conjunto
Sirve para crear un conjunto con todos los caracteres de la cadena.

```python
s = "tres tristes tigres en un trigal"
set(s)
```
