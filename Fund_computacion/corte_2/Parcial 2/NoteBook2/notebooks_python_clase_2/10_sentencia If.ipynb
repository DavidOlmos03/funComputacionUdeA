# Condiciones
## Sentencia **If** (Si)
Permite dividir el flujo de un programa en diferentes caminos por medio de una preginta lógica. El if se ejecuta siempre que la expresión que comprueba devuelva True.

```python
if True:
    print("Si el condicional se cumple se ejecuta esta línea")
    print("y esta también")
```

#### Podemos generar secuencias de if

```python
p = 10
if p == 5:
    print("Este bloque se ejecuta si p=5")
if p == 10:
    print("Este bloque se ejecuta si p=10")
```

#### O también anidar If dentro de If

```python
p = 10
q = 5
if p == 10:
    print("p vale 10")
    if q == 5:
        print("q vale 5")
```

#### Como condición podemos evaluar múltiples expresiones, siempre que éstas devuelvan True o False

```python
if p == 10 and q == 5:
    print("p vale 10 y q vale 5")
```

# Sentencia **Else** (Sino)
Se encadena a un If para ejecutar el caso contrario (en el que no se cumple en la condición).

```python
numero = 11
if numero%2 == 0:
    print("el número es par")
else:
    print("el número es impar")
```

# Sentencia Elif (Sino Si)
Se encadena a un if u otro elif para comprobar múltiples condiciones, siempre que las anteriores no se ejecuten. Se puede usar la sentencia else para establecer una opción predefinida.

```python
instruccion = input("ingrese una instrucción: ")

if instruccion == "iniciar":
    print("Inicia el proceso")
elif instruccion == "continuar":
    print("Continuar proceso")
elif instruccion == "finalizar":
    print("Finalizar proceso")
else:
    print("Opción no válida")
```

**Ejemplo:** Usar elif para dar un criterio de desempeño segun el rango de nota de un estudiante.

```python
nota = float(input("Ingrese una nota entre 0 y 5: "))

if nota >= 4.5:
    print("Gran desempeño")
elif nota >= 4:
    print("Buen desempeño")
elif nota >= 3.5:
    print("Bien, pero puedes mejorar")
elif nota >= 3.0:
    print("Bien, pero tienes la competencia mínima")
else:
    print("No tienes la competencia")
```

#### Es posible simular el funcionamiento de elif con if utilizando expresiones condicionales

```python
nota = float(input("Ingrese una nota entre 0 y 5: "))

if nota >= 4.5:
    print("Gran desempeño")
if nota >= 4.0 and nota < 4.5:
    print("Buen desempeño")
if nota >= 3.5 and nota < 4.0:
    print("Bien, pero puedes mejorar")
if nota >= 3.0 and nota < 3.5:
    print("Bien, pero tienes la competencia mínima")
if nota < 3.0:
    print("No tienes la competencia")
```

# Sentencia **match**

Se implementó despues de pthon 3.10 para recuperar la sentencia **swi** de otros lenguajes de programación basados en C, para evaluar casos.

```python
opcion = input("ingrese una opción: ")

match opcion:
    case "A":
        print("Opicón A")
    case "B":
        print("Opicón B")
    case "C":
        print("Opicón C")
    case _:
        print("Opción predefinida")
```
