
## Sentencia **While** (Mientras)
Se basa en repetir un bloque a partir de evaluar una condición lógica, siempre que ésta sea True.

Queda en las manos del programador decidir el momento en que la condición cambie a False para hacer que el While finalice.


```python
i = 0

while i<=10:
    print("i = ",i)
    i += 1
```

## Sentencia Else en ciclo While
Se encadena al while para ejecutar un bloque de código una vez la condición ya no devuelve True al final del ciclo.

```python
i = 0

while i<=10:
    print("i = ",i)
    i += 1
else:
    print("El valor final de i es ",i)
```

# Leyendo una lista con while. Es más limitada que el For.

```python
numeros = [5,10,15,20,25]
indice = 0

while indice < len(numeros):
    print(numeros[indice])
    indice+=1
```

# Instrucciones **break, continue, pass**
# Sirven para controlar los pasos en un ciclo.

## Instrucción **break**, sirve para "romper" la ejecución de un ciclo en cualquier momento. No se ejecutará el Else, ya que éste sólo se llama al finalizar la iteración.

```python
i = 0

while i<=10:
    print("i = ",i)
    if i == 5:
        print("Rompemos el bucle completamente en i = ",i)
        break
    i += 1
else:
    print("El valor final de i es ",i)
```

## Instrucción **Continue**, sirve para "saltarse" la iteración actual sin romper el bucle.

```python
i = 0

while i<=10:
    if i == 5:
        print("Rompemos el bucle solo en la interaccion i = ",i)
        i +=1
        continue
    print("i = ",i)
    i += 1
else:
    print("El valor final de i es ",i)
```

### **Ejemplo:** Creanenú interactivo

```python
print("Beinvenido a la aplicación")

while True:
    print("""Elija una opción:
    1) Para recibir información
    2) Sumar dos números
    3) Terminar el proceso\n""")
    
    opcion = input("Elija una opción: ")
    
    if opcion == '1':
        print("""Puedes sumar dos números con la opción 2 o terminar el proceso con la opción 3\n""")
    elif opcion == '2':
        print("Suministre dos numeros para sumarlos: ")
        n1 = float(input("Ingrese el primer número: "))
        n2 = float(input("Ingrese el segundo número: "))
        print("La suma es: ",n1+n2,"\n")
    elif opcion == '3':
        print("Proceso terminado")
        break
    else:
        print("Opción no válida, intente de nuevo")
```
