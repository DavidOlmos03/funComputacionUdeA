```python
import numpy as np
import matplotlib.pyplot as plt
import random as rn

def distancia_planeta(e, theta):
    return 1 / (1 + e * np.cos(theta))

def orbitas_reales(e, theta):
    r = distancia_planeta(e, theta)
    x = r * np.sin(theta)
    y = r * np.cos(theta)
    return x, y

theta = np.linspace(0, 2*np.pi, 1000)  #Se genera un arreglo de 1000 números desde el 0 hasta 2pi, que seran los valores que ira tomando el angulo theta

excentricidades = [0, 0.5, 0.8] 

##GRAFICA 1
for e in excentricidades:
    r = distancia_planeta(e, theta)   ##Se calcula la distancia al planeta en función de la excentricidad y el angulo theta, ie r va cambiando cada que theta tiene
                                      #un nuevo valor
    plt.plot(theta, r, label=f'e = {e}')   #se utiliza label para referenciar las curvas trazadas y poderlas identificar en el gráfico 

plt.xlabel('Posición angular (θ)')
plt.ylabel('Distancia al planeta en UA')
plt.title('Distancia al planeta en función de la posición angular')
plt.legend()
plt.grid(True)
plt.show()

#GRAFICA 2
plt.subplot(122)  #Se utiliza para controlar la vista de la grafica
for e in excentricidades:
    x, y = orbitas_reales(e, theta)
    plt.plot(x, y, label=f'e = {e}')

plt.xlabel('x(θ)')
plt.ylabel('y(θ)')
plt.title('Órbitas reales del planeta')
plt.legend()
plt.grid(True)

#plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def calcular_posicion(x0, y0, v0x, v0y):
    g = 9.8  # aceleracion de la gravedad (m/s^2) en un campo gravitacional cte 
    
    t_max = 2 * v0y / g   #Tiempo maximo (antes de que la particula toque el suelo)
    
    t = np.linspace(0, t_max, num=100)  # Creamos un arreglo de tiempo desde 0 hasta t_max
    
    # Calculamos las posiciones en x y y para cada t
    x = x0 + v0x * t
    y = y0 + v0y * t - 0.5 * g * t**2

    return x, y  

#Caso a
x_a, y_a = calcular_posicion(0, 100, 0, 0)

#Caso b
x_b, y_b = calcular_posicion(0, 100, 1, 0)

#Caso c
x_c, y_c = calcular_posicion(0, 0, 0, 10)

#Caso d
x_d, y_d = calcular_posicion(0, 0, 1, 10)

# Graficar las trayectorias
#plt.figure(figsize=(10, 8))
#plt.title('Trayectorias de partículas bajo campo gravitacional constante')
#plt.grid(True)
plt.xlabel('Posición en x (m)')
plt.ylabel('Posición en y (m)')
plt.plot(x_a, y_a, label='Caso a')
plt.show()
plt.xlabel('Posición en x (m)')              ###ARREGLAR 
plt.ylabel('Posición en y (m)')
plt.plot(x_b, y_b, label='Caso b')
plt.show()
plt.xlabel('Posición en x (m)')
plt.ylabel('Posición en y (m)')
plt.plot(x_c, y_c, label='Caso c')
plt.show()
plt.xlabel('Posición en x (m)')
plt.ylabel('Posición en y (m)')
plt.plot(x_d, y_d, label='Caso d')
plt.show()


#plt.legend()


```
