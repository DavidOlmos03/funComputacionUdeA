```python
import matplotlib.pyplot as plt
import numpy as np
from math import *

x = np.arange(-10,10,5)
num_segmentos = 300
angulo = np.linspace(0, 2*np.pi, num_segmentos)
##r = (1-np.exp(1)**2)/(1+np.exp(1)*np.cos(angulo))
ax = (np.cos(angulo))**2 
ay = (np.sin(angulo))**2
y = 1 / np.sqrt(ax/5 + ay/3)

plt.plot(y)
plt.plot(0,0,"o")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Proof graphics')
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig1 = plt.figure()
ax1 = fig1.add_subplot(111)
ax1.add_patch(
    patches.Ellipse(
        (0.5, 0.5),   # (x,y)
        0.6,          # width
        0.5,          # height
    )
)

# fig1.savefig('ellipse1.png', dpi=90, bbox_inches='tight')
plt.show()
```
